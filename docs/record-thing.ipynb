```python
from collections import namedtuple
from record_thing.taxonomy.taxonomy_parser import root

product_type_tuple = namedtuple('product_type', ['rootName', 'name', 'url', 'gpcRoot', 'gpcName', 'gpcCode', 'unspscID'])

def gen_from_node(node, base = []):
    base_names = ">".join(base)
    yield product_type_tuple(base_names, node['name'], None, base_names, node['name'], node['code'], None)
    for child in node['children']:
        yield from gen_from_node(child, base + [node['name']])

def generate_product_type():
    for child in root['children']:
        yield from gen_from_node(child, [])

lst = [l for l in generate_product_type()]
lst
```
